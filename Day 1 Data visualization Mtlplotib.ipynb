```python
df = pd.read_csv('student.csv')
```

```python
plt.title('Roll vs Marks')
plt.xlabel('Roll Numbers')
plt.ylabel('Marks')
plt.grid()
plt.xlim(0,11)
plt.ylim(40,100)
plt.plot(df['roll'],df['marks'], color= 'red',marker= 'o',mfc='green',lw=4)
```

```python
plt.title('My First plot')
plt.xlabel('First Value')
plt.ylabel('second Value')
plt.grid()
plt.xlim(0,70)
plt.ylim(0,100)
plt.yticks(range(0,100,10))
plt.plot(x, y, color = 'red', label = '2023')
plt.plot(x, z, color = 'green', label ='2024')

plt.legend()
```
