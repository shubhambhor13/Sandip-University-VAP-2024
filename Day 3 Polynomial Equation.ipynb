```python
import numpy as np
import matplotlib.pyplot as plt
```

###  Linear Equation

```python
x = np.arange(0,10,0.25)
```

```python
x
```

```python
y = x*4.23 + 36.07
```

```python
plt.plot(x,y)
```

### Polynomial Equation

```python
y = x**2*4.23 + 36.07
```

```python
y
```

```python
plt.plot(x,y)
```
