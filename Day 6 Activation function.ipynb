```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5,5,0.25)
```

# Sigmoid function

```python
def sigmoid(z):
    val = 1/(1 + np.exp(-z))
    return val
```

```python
plt.plot(x, sigmoid(x))
```

# ReLU

```python
def relu(z):
    if z > 0:
        return z
    else:
        return 0
```

```python
y = [relu(n) for n in x]
```

```python
plt.plot(x, y)
```

# Softmax

```python
def softmax(z):
    val = np.exp(z) / np.sum(np.exp(z))
    return val
```

```python
num = [1.3, 1.25, 1.5, 1.04, 0.94]
```

```python
softmax(num)
```

## Tanh Function

```python
plt.plot(x, np.tanh(x))
```

```python
plt.grid()
plt.plot(x, np.tanh(x), color ='red', label ='tanh')
plt.plot(x, sigmoid(x), color ='blue', label ='sigmoid')
plt.legend();
```
